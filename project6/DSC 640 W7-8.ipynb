```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("birth-rates-yearly.csv")
df
```

```python
avg = df.groupby('year')[['rate']].mean().reset_index()
avg
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(avg['year'], avg['rate'])           # scatter plot showing actual data
plt.xlabel('Year')
plt.ylabel('Birth Rate')
plt.title('Yearly Birth Rate')
plt.show()
```

```python
df1=pd.read_csv('crimerates-by-state-2005.csv')
df1=df1.drop([0,9])
df1
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(df1['population'], df1['murder'], df1['robbery'], alpha=0.5)           # scatter plot showing actual data
plt.xlabel('Population')
plt.ylabel('Murder Rate')
plt.title('Murder rate per population with robbery rates as bubble size')
plt.show()
```
