```python
df3=pd.read_csv("mlst3.csv")
df3.columns=['Overall Standing','Division Standing','Team','Pct']
df3
##import team win percentage
```

```python
df2=pd.read_csv("mlst2.csv")
df2.columns=['Leaderboard','Player', 'Team','AVG']
df2
##Import Hitters
```

```python
team_df='https://en.wikipedia.org/wiki/Wikipedia:WikiProject_Baseball/Team_abbreviations'
team=pd.read_html(team_df)##receive data and assign to a dataframe
team=team[0]
team.columns=['abv','full']
team
##import table to indicate the change between abbreviations and full name
##change to full name so it can be compatible to other data
```

```python
df2.replace(['ANA','ARI','ATL','BAL','BOS','CHC','COL','WAS','HOU','NYN','CIN','DET','LAA','LAN','CLE','PIT','TBA','SLN','MIL','MIA','OAK','NYA','TEX','KCA','TOR','SEA','MIN','CHA'],['Anaheim Angels','Arizona Diamondbacks','Atlanta Braves','Baltimore Orioles','Boston Red Sox','Chicago Cubs','Colorado Rockies','Washington Nationals','Houston Astros','New York Mets','Cincinnati Reds','Detroit Tigers','Los Angeles Angels','Los Angeles Dodgers','Cleveland Indians','Pittsburgh Pirates','Tampa Bay Rays','St. Louis Cardinals','Milwaukee Brewers','Miami Marlins','Oakland Athletics','New York Yankees','Texas Rangers','Kansas City Royals','Toronto Blue Jays','Seattle Mariners','Minnesota Twins','Chicago White Sox'], inplace=True)
df2
##replace ABV with full team names
```

```python
##convert CSV to SQL for batters
```

```python
conn= sqlite3.connect("mile2")
```

```python
c=conn.cursor()
```

```python
c.execute('CREATE TABLE IF NOT EXISTS bat (Leaderboard,Player,Teams,AVG)')
conn.commit()
```

```python
df2.to_sql('bat', conn, if_exists='replace', index = False)
```

```python
c.execute('''  SELECT * FROM bat''')
for row in c.fetchall():
    print (row)
```

```python
conn3= sqlite3.connect("mile3")
c3=conn3.cursor()
c3.execute('CREATE TABLE IF NOT EXISTS standing (Overall_Standing,Division_Standing,Teams,Pct)')
conn3.commit()
df3.to_sql('standing', conn3, if_exists='replace', index = False)
c3.execute('''  SELECT * FROM standing''')
for row in c3.fetchall():
    print (row)
    
##convert CSV to SQL for team winning percentage
```

```python
conn4= sqlite3.connect("mile4")
c4=conn4.cursor()
c4.execute('CREATE TABLE IF NOT EXISTS pitch (Leaderboard,Name,Team,ERA)')
conn4.commit()
df4.to_sql('pitch', conn4, if_exists='replace', index = False)
c4.execute('''  SELECT * FROM pitch''')
for row in c4.fetchall():
    print (row)
    
##convert CSV to SQL for Pitchers
```

```python
#Join the datasets together in Python into 1 dataset
df=pd.concat([df4, df2, df3], axis=0)
df
```

```python
conn1= sqlite3.connect("comb")
c1=conn1.cursor()
c1.execute('CREATE TABLE IF NOT EXISTS combine (Leaderboard,Name,Teams,ERA,Player,AVG,Overall_Standing,Division_Standing,Pct)')
conn1.commit()
df.to_sql('combine', conn1, if_exists='replace', index = False)
c1.execute('''SELECT * FROM combine''')
for row in c1.fetchall():
    print (row)
    
##convert CSV to SQL for full database
```

```python
cnt= df.groupby('Team').count()
cnt
##finds counts for each team in top players
##ERA count indicates top pitchers
##AVG count indicates top hitters
##Leaderboard indicated all top players
```

```python
f = plt.figure()
f.set_figwidth(40)
f.set_figheight(15)
plt.bar(cnt.index,cnt['Leaderboard'])
plt.xlabel('Team')
plt.ylabel('Count')
plt.title('How many top 50 players each team has')
plt.show() 

##overall number of top players each team has in the form of a Bar graph
```

```python
plt.scatter(cnt['AVG'],cnt['ERA'])
plt.xlabel('Hitters in top 50')
plt.ylabel('Pitchers in top 50')
plt.title('Compare number of hitters and pitchers in the top 50s')
plt.show() 
##plots scatter plot for all top players for each team, comparing batters and hitters
```
