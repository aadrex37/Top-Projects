```python
##Baseball was very different than it was a long time ago
##I am going to remove all entries prior to 1985, which limits to the modern age and the start of the salary data
##Most say the start is at 1901, but statistics were very different there
##article where this is discussed
##To determine the best predictive statistics for batters, I will use weighted On base percentage
##This is the best statistics that I can find given the information to which I have access
##https://www.espn.com/mlb/columns/story?columnist=stark_jayson&id=2471349
##https://library.fangraphs.com/the-beginners-guide-to-deriving-woba/
##For Pitching, I will be comparing ERA to Fielding independent pitching
##https://legacy.baseballprospectus.com/glossary/index.php?search=FIP
```

```python
%matplotlib inline
```

```python
from yellowbrick.features import Rank2D
```

```python
woba = ['playerID', 'yearID','AB','H','2B','3B','HR','BB','IBB','HBP','SF']
fip = ['playerID', 'yearID','ERA','HR','HBP','BB','SO','IPouts']
```

```python
x = batting[woba]
y= pitching[fip]
##Creating new DF with just columns of significance
```

```python
fip1 = (((13*y['HR'])+(3*(y['HBP']+y['BB']))-(2*y['SO']))/(y['IPouts']/3))+3.2
ubb = (x['BB']-x['IBB'])
single =x['H']-x['2B']-x['3B']-x['HR']
woba1 = (0.69*ubb + 0.72*x['HBP']+0.89* single + 1.27* x['2B'] + 1.62*x['3B'] +2.1 * x['HR'])/(x['AB']+ubb+x['SF']+x['HBP'])
##Calculating advanced statistics
```

```python
dfb = x[(x['yearID']>=1985) & (x['AB']>=100)]
dfb
##Removing all entrys prior to 1985 and players with less than 100 at bats
##requires 100 at bats to become statistically stable and removes small-sample bias
```

```python
dfp = y[(y['yearID']>=1985) & (y['IPouts']>=60)]
dfp 
##Removing all entrys prior to 1985 and requires at least 20 innings pitched (small sample bias)
```

```python
dfb['wOBA']= woba1
dfb.head(10)
##adding wOBA to df
```

```python
dfp['FIP']= fip1
dfp.head(10)
##Adding FIP to new Data Frame
```

```python
catg = ['playerID','wOBA','yearID']
dfb1 = dfb[catg]
dfb1
##creating a simpler df with just important information
```

```python
catgp = ['playerID','ERA','FIP','yearID']
dfp1 = dfp[catgp]
dfp1
##creating a simpler df with just important information
```

```python
import statsmodels.formula.api as smf
```

```python
# Initialise and fit linear regression model using `statsmodels`
model = smf.ols('wOBA ~ salary', data=new_df1)
model = model.fit()
```

```python
# Predict values
pred = model.predict()

# Plot regression against actual data
plt.figure(figsize=(12, 6))
plt.plot(new_df1['salary'], new_df1['wOBA'], 'o')           # scatter plot showing actual data
plt.plot(new_df1['salary'], pred, 'r', linewidth=2)   # regression line
plt.xlabel('Salary')
plt.ylabel('Weighted On Base Average')
plt.title('wOBA v Salary')
plt.show()
```
