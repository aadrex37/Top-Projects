```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df = pd.read_csv("w11-12/education.csv")
df
```

```python
math=df['math']
math
```

```python
plt.figure(figsize=(15, 6))
plt.hist(math, bins=20)
plt.title('State Math Scores')
plt.xlabel('Math Scores')
plt.ylabel('Number of States')
plt.show()
```

```python
plt.figure(figsize=(15, 6))
plt.boxplot(math)
plt.title('State Math Scores')
plt.ylabel('Math scores')
plt.show()
```

```python
palette = sns.color_palette("Blues_r", len(math))
```

```python
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.set_yticks([1])
ax.set_yticklabels(['Math Scores'])

prev_limit = 0
for idx, lim in enumerate(math):
    ax.barh([1], lim-prev_limit, left=prev_limit, height=15, color=palette[idx])
    prev_limit = lim
```

```python
read= df['reading']
write=df['writing']
```

```python
plt.figure(figsize=(15, 6))
plt.scatter(math,read,c='b',label='Math Scores Versus Reading Scores')
plt.scatter(math,write,c='r',label='Math Scores Versus Writing Scores')
plt.title('State Educational Scores')
plt.xlabel('Math Scores')
plt.ylabel('Reading and Writing Scores')
plt.show()
```
